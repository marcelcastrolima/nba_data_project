```python
from nba_api.stats.static import players

# Find players by full name.
players.find_players_by_first_name('lebron')

```

```python

lebron = playercareerstats.PlayerCareerStats(player_id='2544') 

# pandas data frames (optional: pip install pandas)
lebron = lebron.get_data_frames()[0]

lebron
```

```python
lebron['PPG'] = lebron['PTS']/lebron['GP']
lebron['MINPG'] = lebron['MIN']/lebron['GP']
```

```python
lebron
```

```python
lebron = lebron.rename(columns={"TEAM_ID": "TeamID"})
```

```python
lebron
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
```

```python
plt.figure(figsize=(16,4))
matplotlib.rc('xtick', labelsize=8) 
barplot = sns.barplot(data=lebron, x="SEASON_ID", y="PPG")

for i in barplot.containers:
    barplot.bar_label(i,)

plt.show()
```

```python
plt.figure(figsize=(16,4))
matplotlib.rc('xtick', labelsize=8) 
barplot = sns.barplot(data=lebron, x="SEASON_ID", y="MINPG")

for i in barplot.containers:
    barplot.bar_label(i,)

plt.show()
```
