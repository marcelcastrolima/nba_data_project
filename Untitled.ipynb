```python
standings_multiple_years = pd.DataFrame()
```

```python
for year in range(2015, 2020):
    standings = leaguestandings.LeagueStandings(league_id="00", season=str(year), season_type="Regular Season")
    standings = standings.get_data_frames()[0]
    print(standings)
    standings_multiple_years = standings_multiple_years.append(standings, ignore_index=True)
    time.sleep(2)
```

```python
standings_multiple_years['SeasonID'] = standings_multiple_years['SeasonID'].str[1:]
```

```python
standings_multiple_years
```

```python
standings_multiple_years = standings_multiple_years.rename(columns={"TeamID": "TEAM_ID", "SeasonID": "YEAR"})
```

```python
standings_multiple_years
```

```python
result_all_teams = pd.DataFrame()
```

```python
for team in pd.unique(standings_multiple_years['TEAM_ID']):

    by_team = teamyearbyyearstats.TeamYearByYearStats(team_id=team, per_mode_simple='PerGame', season_type_all_star="Regular Season") 
    by_team = by_team.get_data_frames()[0]
    by_team['YEAR'] = by_team['YEAR'].str[:4]
    by_team['AST_TOV_RAT'] = by_team['AST']/by_team['TOV']
    by_team = by_team.drop(['AST', 'TOV'], axis=1)
    by_team = by_team[['TEAM_ID', 'TEAM_NAME', 'YEAR', 'WIN_PCT', 'FG_PCT', 'REB', 'AST_TOV_RAT', 'PTS']]
    print(team)
    
    result_all_teams = result_all_teams.append(by_team, ignore_index=True)
    time.sleep(2)
```

```python
result_all_teams['YEAR'] =result_all_teams['YEAR'].astype(int)
```
